```python
# Dependencies
import requests
import json
import quandl
import pandas as pd
from datetime import datetime
import plotly.express as px
```

```python
Final_df = pd.read_csv('../../db/final_complete_merged_dataset.csv',sep=',')
Final_df.head()
```

```python
vader_group = Final_df.groupby(['call_section'])
vader_bars = pd.DataFrame(vader_group['vader_sentiment'].mean())
our_bars = pd.DataFrame(vader_group['model_results'].mean())
vader_bars
our_bars
```

```python
group_df = Final_df.loc[Final_df['call_section']=='presentation']
group_df = group_df.groupby(['ticker'])
ticker_unique = group_df.ticker.unique()
group_sent_avg = group_df['model_results'].mean()
group_price_avg = group_df['percent_change_price'].mean()
group_vol_avg = abs(group_df['percent_change_volume'].mean())
group_cap_avg = group_df['marketCap'].mean()
```

```python
fig = px.scatter(x=group_sent_avg, y=group_price_avg, trendline="ols",hover_name=ticker_unique, size=group_vol_avg)
fig.show()
```

```python
fig = px.scatter(x=group_sent_avg, y=group_vol_avg, trendline="ols",hover_name=ticker_unique, size=group_vol_avg)
fig.show()
```

```python
vader_bars=vader_bars.reset_index()
fig2 = px.bar(vader_bars,x='call_section', y='vader_sentiment')
fig2.show()
```

```python
our_bars=our_bars.reset_index()
fig2 = px.bar(our_bars,x='call_section', y='model_results')
fig2.show()
```
