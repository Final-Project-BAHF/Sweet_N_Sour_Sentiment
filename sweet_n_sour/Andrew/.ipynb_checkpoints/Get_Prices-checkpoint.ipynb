```python
# Dependencies
import pandas as pd
import pandas_datareader as web
import datetime
```

```python
start = datetime.datetime(2019, 1, 1)
end = datetime.datetime(2019, 11, 14)
tickers = ['IVZ','GM','UAA','WFC','TWTR','SBUX','CRM','AMZN','GE','CMG']
df_output = pd.DataFrame()
for ticker in tickers:
    mydata = web.DataReader(ticker, 'yahoo', start, end)
    df = pd.DataFrame(mydata)
    df['ticker'] = ticker
    df_output = df_output.append(df)
df_output

```

```python
df_output.to_csv('../../db/ticker_closing_prices.csv', sep=',')


```
